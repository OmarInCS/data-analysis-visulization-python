## Clean

```python
patients_clean = patients.copy()
treatments_clean = treatments.copy()
reactions_clean = reactions.copy()
```

#### Define:
   - change datatypes of `zIp_code` and `birthdate` to str and datetime

##### Code:

```python
patients_clean["zip_code"] = patients_clean["zip_code"].astype(str).str[:-2]
patients_clean["birthdate"] = pd.to_datetime(patients_clean["birthdate"])
```

##### Test:

```python
patients_clean.dtypes
```

#### Define:
   - separate email and mobile in `contact` column to two columns

##### Code:

```python
pattern = r"(\d{3}-\d{3}-\d{4}|\d{10}|1 \d{3} \d{3} \d{4}|\+1 \(\d{3}\) \d{3}-\d{4})"
patients_clean["mobile"] = patients_clean["contact"].str.extract(pattern)
patients_clean["email"] = patients_clean["contact"].str.replace(pattern, "", regex=True)
```

##### Test:

```python
patients_clean[["contact", "mobile", "email"]].sample(10)
```

#### Define:
   - drop duplicated rows of 'John Doe'
   - drop duplicated rows for patients with IDs 30, 503, 283

##### Code & Test:

```python
patients_clean.drop_duplicates(subset=["given_name", "surname"], inplace=True)
cond = patients_clean["patient_id"].isin([30, 503, 283])
patients_clean = patients_clean[~cond]
```

```python
patients_clean.duplicated(subset=["given_name", "surname"]).sum()
```

```python
mask = patients_clean.dropna(subset=["contact"]).duplicated(subset=["contact"], keep=False)
patients_clean.dropna(subset=["contact"])[mask]
```

#### Define:
   - convert `weight` and `height` to kg and cm

```python
patients_clean.loc[:, "weight"] = np.round(patients.loc[:, "weight"] * 0.453592, 2)
patients_clean.loc[:, "height"] = patients.loc[:, "height"] * 2.54
```

```python
patients_clean[["weight", "height"]].sample(10)
```

#### Define:
  - recalculate `weight` for patient with ID 211
  - recalculate `height` for patient with ID 5
  - recalculate `bmi`

```python
cond = patients_clean["patient_id"] == 211
patients_clean.loc[cond, "weight"] = np.round(patients_clean.loc[cond, "bmi"] * (patients_clean.loc[cond, "height"] / 100) ** 2, 2)
```

```python
cond = patients_clean["patient_id"] == 211
patients_clean[cond]
```

```python
cond = patients_clean["patient_id"] == 5
patients_clean.loc[cond, "height"] = 183
```

```python
patients_clean[cond]
```

```python
patients_clean["bmi"] = np.round(patients_clean["weight"] / (patients_clean.loc[cond, "height"] / 100) ** 2, 2)
```

#### Define:
   - recalculate the `hba1c_change` column

```python
treatments_clean["hba1c_change"] = treatments_clean["hba1c_start"] - treatments_clean["hba1c_end"]
```

```python
# test
treatments_clean["hba1c_change"].isnull().sum()
```

#### Define:
   - create one column for dose and another for treatment type
   - separate values in columns `auralin` and `novadra`

```python
treatments_clean = treatments_clean.melt(id_vars=["given_name", "surname", "hba1c_start", "hba1c_end", "hba1c_change"],
                    value_vars=["auralin", "novodra"]
                     , var_name="treatment", value_name="dose").replace("-", np.nan).dropna()
```

```python
# test
treatments_clean.sample(10)
```

```python
temp = treatments_clean["dose"].str.extract("(\d+)u - (\d+)u").rename(columns={0: "start_dose", 1: "end_dose"})
treatments_clean = treatments_clean.join(temp).drop(columns="dose")
```
