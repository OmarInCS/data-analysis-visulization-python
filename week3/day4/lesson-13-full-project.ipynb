## Clean

```python
patients_clean = patients.copy()
treatments_clean = treatments.copy()
reactions_clean = reactions.copy()
```

#### Define
   - change given_name and surname to title case in tables `treatments` and `adverse_reactions`, then join the tables

#### Code

```python
treatments_clean["given_name"] = treatments_clean["given_name"].str.title()
treatments_clean["surname"] = treatments_clean["surname"].str.title()

reactions_clean["given_name"] = reactions_clean["given_name"].str.title()
reactions_clean["surname"] = reactions_clean["surname"].str.title()

treatments_clean = treatments_clean.merge(reactions_clean, on=["given_name", "surname"], how="outer")
```

#### Test

```python
treatments_clean
```

#### Define
 - create new column for treatment type, that contain 'auralin' or 'novodra'


#### Code

```python
treatments_clean = treatments_clean.melt(id_vars=["given_name", "surname", "hba1c_start", "hba1c_end", "hba1c_change", "adverse_reaction"], 
                                          value_vars=["auralin", "novodra"], var_name="treatment", value_name="dose")

treatments_clean["dose"] = treatments_clean["dose"].replace("-", np.nan)
treatments_clean.dropna(subset=["dose"], inplace=True)
```

#### Test

```python
treatments_clean
```

#### Define
   - separate the **dose** column into two columns **start_dose** and **end_dose**

#### Code

```python
treatments_clean[["start_dose", "end_dose"]] = treatments_clean["dose"].str.extract("(\d+)\D+(\d+)")
treatments_clean.drop(columns=["dose"], inplace=True)
```

#### Test

```python
treatments_clean
```

#### Define
   - calculate **hba1c_change** using the equation **hba1c_start** - **hba1c_end**
   - fill *NaN* in **adverse_reaction** with 'no reaction'

#### Code

```python
treatments_clean["hba1c_change"] = treatments_clean["hba1c_start"] - treatments_clean["hba1c_end"]
treatments_clean["adverse_reaction"] = treatments_clean["adverse_reaction"].fillna("no reaction")
```

#### Test

```python
treatments_clean
```

#### Define
   - replace all state's abbreviation with the full name

#### Code

```python
state_abbrev_map = pd.read_html("https://www.ssa.gov/international/coc-docs/states.html")[0].set_index(1).to_dict()[0]

def map_abbrev_tofull_name(abbrev):
    if abbrev in state_abbrev_map:
        return state_abbrev_map[abbrev]
    elif abbrev is not np.nan:
        return abbrev.upper()

patients_clean["state"] = patients_clean["state"].map(map_abbrev_tofull_name)
```

#### Test

```python
patients_clean["state"].value_counts()
```

#### Define
   - make two columns **email** and **phone**, out off the **contact** column

#### Code

```python
patients_clean["phone"] = patients_clean["contact"].str.extract("(\+1 \(\d+\) \d+-\d+|\d+-\d+-\d+|1 \d+ \d+ \d+)")
patients_clean["email"] = patients_clean["contact"].str.replace("(\+1 \(\d+\) \d+-\d+|\d+-\d+-\d+|1 \d+ \d+ \d+)", "")
patients_clean.drop(columns=["contact"], inplace=True)
```

#### Test

```python
patients_clean.sample(5)
```

#### Define
   - convert the type of **zip_code** column to string
   - convert the type of **birthdate** column to datetime

#### Code


```python
patients_clean["zip_code"] = patients_clean["zip_code"].astype(str)
patients_clean["birthdate"] = pd.to_datetime(patients_clean["birthdate"], format="%m/%d/%Y")
```

#### Test

```python
patients_clean.info()
```

#### Define
   - remove 'Jake Jakobsen' and 'Sandy Taylor'
   - remove duplicated rows
   

#### Code

```python
mask = ((patients_clean["given_name"] == "Jake") & (patients_clean["surname"] == "Jakobsen")) | ((patients_clean["given_name"] == "Sandy") & (patients_clean["surname"] == "Taylor"))
patients_clean = patients_clean[~mask]
patients_clean.drop_duplicates(subset=["address"], inplace=True)
```

#### Test

```python
patients_clean.duplicated(subset="address").sum()
```

#### Define
   - convert the weight to *Kg* and the height to *m*
   - correct weight of 'Camilla Zaitseva'
   - correct height of 'Tim Neudorf'

#### Code

```python
# bmi = weight / height^2
# height = (weight / bmi)^0.5
# weight = bmi * height^2
patients_clean["weight"] = patients_clean["weight"] / 2.205
patients_clean["height"] = patients_clean["height"] / 39.37

mask = ((patients_clean["given_name"] == "Camilla") & (patients_clean["surname"] == "Zaitseva"))
patients_clean.loc[mask, "weight"] = patients_clean.loc[mask, "bmi"] * patients_clean.loc[mask, "height"] ** 2

mask = ((patients_clean["given_name"] == "Tim") & (patients_clean["surname"] == "Neudorf"))
patients_clean.loc[mask, "height"] = (patients_clean.loc[mask, "weight"] / patients_clean.loc[mask, "bmi"]) ** 0.5
```

#### Test

```python
mask = ((patients_clean["given_name"] == "Camilla") & (patients_clean["surname"] == "Zaitseva")) | ((patients_clean["given_name"] == "Tim") & (patients_clean["surname"] == "Neudorf"))
patients_clean[mask]
```

## Model


```python
model_columns = ['assigned_sex', 'birthdate', 'weight', 'height', 'hba1c_start', 'hba1c_change', 'treatment', 'start_dose']
model_data = patients_full[model_columns].copy().dropna()
model_data
```

```python
from datetime import datetime


model_data["age"] = model_data.birthdate.map(lambda dob: round((datetime.now() - dob).days / 365))
model_data.drop(columns=["birthdate"], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = model_data.drop(columns=["hba1c_change"])
y = model_data["hba1c_change"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import LabelEncoder

lb_sex = LabelEncoder()
X_train.loc[:, "assigned_sex"] = lb_sex.fit_transform(X_train["assigned_sex"])
X_test.loc[:, "assigned_sex"] =  lb_sex.transform(X_test["assigned_sex"])

lb_treat = LabelEncoder()
X_train.loc[:, "treatment"] = lb_treat.fit_transform(X_train["treatment"])
X_test.loc[:, "treatment"] =  lb_treat.transform(X_test["treatment"])
X_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("Training Score: ", round(model.score(X_train, y_train) * 100, 2))
print("Test Score:", round(model.score(X_test, y_test) * 100, 2))
```
